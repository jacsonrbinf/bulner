```python
# parameters

bugdata = 'dados-brutos-tomcat.csv'
embedding_model = 'stack_overflow.embedding.vec'
```

```python
import pandas as pd
import numpy as np

# reading CSV bug reports patterns 
df = pd.read_csv(bugdata)
df
```

```python
# data discretization from numeric complexity measures
code_measures = list(df.columns[4:44])
labels = ['lv0','lv1','lv2','lv3','lv4','lv5','lv6','lv7','lv8','lv9']
for measure in code_measures:
    print(measure)
    c = pd.cut(df[measure],10,labels=labels)
    df[measure] = c 

df
```

```python
import os
# network building

network = {}

# bug_id <-> terms
for index, row in df.iterrows():
    bug_id = row['bug_id']
    terms = (str(row['summary_stemmed'])+" "+str(row['description_stemmed'])).split(' ')
    for term in terms:
        network[str(bug_id)+":bug\t"+term+":term"]=1
        
# bug_id <-> filename
for index, row in df.iterrows():
    bug_id = row['bug_id']
    filename = row['Name']
    filename = "["+str(bug_id)+"] "+filename
    network[str(bug_id)+":bug\t"+filename+":file"]=1
    #for term in terms:
        

# filename <-> code_measures
code_measures = list(df.columns[4:44])
for index, row in df.iterrows():
    bug_id = row['bug_id']
    filename = row['Name']
    node1 = "["+str(bug_id)+"] "+filename
    for measure in code_measures:
        measure_value = row[measure]
        node2 = measure+"_"+str(measure_value)
        network[node1+":file\t"+node2+":measure"]=1


if os.path.exists(bugdata+".network"):
  os.remove(bugdata+".network")
        
for edge in network:
    with open(bugdata+".network", "a") as myfile:
        myfile.write(edge+"\t1\n")
```

```python
# generating label data (from external embedding model) for network regularization 
from gensim.models.keyedvectors import KeyedVectors
word_vect = KeyedVectors.load_word2vec_format(embedding_model, binary=False)
```

```python
# testing embedding model (getting similar/semantic words)
word_vect.most_similar('abort')
```

```python
# getting terms for regularization
terms_list = {}
for index, row in df.iterrows():
    terms = (str(row['summary_stemmed'])+" "+str(row['description_stemmed'])).split(' ')
    for term in terms:
        terms_list[term]=1

        
if os.path.exists(bugdata+".labels"):
    os.remove(bugdata+".labels")
        
for term in terms_list:
    if term in word_vect:
        f = word_vect[term]
        y = ""
        for v in f:
            y += str(v)+","
        y += "#"
        with open(bugdata+".labels", "a") as myfile:
            myfile.write(term+':term'+"\t"+y.replace(',#','')+"\n")        

print('Generaring data for regularization: OK')
```
